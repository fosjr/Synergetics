```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# define the initial system state (x, y, z positions in space)
initial_state = [0.1, 1.0, 0.1]

# define the system parameters sigma, rho, and beta
a = 0.2 #0.1, 0.3
b = 0.2 #0.1, 0.4
c = 4.3 #2.6, 3.5, 4.1, 4.23, 4.3, 4.6, 5.7, 8.5, 14

# define the time points to solve for, evenly spaced between the start and end times
start_time = 0
end_time = 60
interval = 400
tspan = np.linspace(start_time, end_time, end_time * interval)
```

```python
# define the roessler system
def roessler_system(current_state, t, a, b, c):
    x, y, z = current_state
    dx_dt = - y - z
    dy_dt = x + a * y
    dz_dt = b + z * (x - c)
    return [dx_dt, dy_dt, dz_dt]
```

```python
# get the points to plot, by integrating the system of equations
points = odeint(roessler_system, initial_state, tspan, args=(a, b, c))
```

```python
# draw plots
fig = plt.figure(figsize=(12, 9))
ax = plt.axes(facecolor='black') 
plt.plot(points[:, 0], points[:, 1], 'r', linewidth = 1.8, label='a= ' + str(a) + '\nb= ' + str(b) + '\nc= ' + str(c))
plt.title("Roessler system")
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

fig = plt.figure(figsize=(12, 9))
ax = plt.axes(facecolor='black') 
plt.plot(points[:, 0], points[:, 2], 'r', linewidth = 1.8, label='a= ' + str(a) + '\nb= ' + str(b) + '\nc= ' + str(c))
plt.title("Roessler system")
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('z')
plt.grid()
plt.show()

fig = plt.figure(figsize=(12, 9))
ax = plt.axes(facecolor='black') 
plt.plot(points[:, 1], points[:, 2], 'r', linewidth = 1.8, label='a= ' + str(a) + '\nb= ' + str(b) + '\nc= ' + str(c))
plt.title("Roessler system")
plt.legend(loc='best')
plt.xlabel('y')
plt.ylabel('z')
plt.grid()
plt.show()
```
