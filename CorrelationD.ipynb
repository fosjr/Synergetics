```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
x = np.array([[0, 0], [1, 0], [0.5, 0.866]]) 

y = x
temp = y[0]
```

```python
n = len(x)

for i in range(50000):
    cur = x[np.random.randint(n)] #Random point from x
    temp = (temp + cur)/(n - 1)
    y = np.append(y, temp.reshape(1, 2), axis=0)
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
plt.plot(y[:, 0], y[:, 1], '.', markersize=0.5)
plt.show()
```

```python
print("y:", y)
```

```python
def Correlation_dimension(fnp):
    length = len(fnp)   #num of points
    c_dim = -1          #dimension
    eps = 0.005
    g_eps = 0         #num of pairs of nearest points (< eps)

    for i in range(length):         #very bad cycle :(
        for j in range(i, length):
            if(math.sqrt( (fnp[i][0] - fnp[j][0])**2 + (fnp[i][1] - fnp[j][1])**2 ) < eps):
                g_eps += 1

    c_dim = np.log(g_eps/(length**2))/np.log(eps)

    return c_dim
```

```python
Cdim = Correlation_dimension(y)
print("Correlation dimension: ", Cdim)
```
